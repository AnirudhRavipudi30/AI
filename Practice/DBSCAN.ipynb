```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

file_path = '/Users/anirudhravipudi/Desktop/AI/Practice/customer_segmentation_data.csv'
df = pd.read_csv(file_path)

features = df[['Age', 'Income Level', 'Coverage Amount', 'Premium Amount']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# 🚀 Apply DBSCAN
dbscan = DBSCAN(eps=0.15, min_samples=10)# you can tune eps and min_samples
clusters = dbscan.fit_predict(scaled_features)

# 🏷️ Assign cluster labels
df['DBSCAN_Cluster'] = clusters

# 📊 Visualize (2D projection: Income vs. Coverage)
plt.figure(figsize=(8, 6))
plt.scatter(df['Income Level'], df['Coverage Amount'], 
            c=df['DBSCAN_Cluster'], cmap='plasma', s=50)
plt.xlabel('Income Level')
plt.ylabel('Coverage Amount')
plt.title('DBSCAN Clustering (eps=0.5, min_samples=5)')
plt.grid(True)
plt.show()

# 🧾 Check number of clusters and noise
print("Unique Clusters:", set(clusters))
print("Noise points (-1):", list(clusters).count(-1))
```
